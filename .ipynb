### split feature and target variable

```python
target = cancer.diagnosis
feature = cancer.drop('diagnosis' ,axis = 1)
feature.head()
```

```python
target.head()
```

```python
features = (feature - feature.mean())/(feature.max()-feature.min())
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 6)
pca.fit(features)

final_data = pca.transform(features)
final_data = pd.DataFrame(final_data,columns = ['Dimension 1','Dimension 2','Dimension 3',\
                                                 'Dimension 4','Dimension 5','Dimension 6'])
final_data.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.model_selection import KFold
```

```python
clf1 = DecisionTreeClassifier()
clf2 =  RandomForestClassifier()
clf3 = LogisticRegression()
kf = KFold(n_splits = 5)

#split the data and train data with kfolds
for train_index , test_index in kf.split(final_data):
    x_train , x_test = final_data.iloc[train_index] , final_data.iloc[test_index]
    y_train , y_test = target.iloc[train_index],target.iloc[test_index]
    
    #fit the model and predict
    clf1.fit(x_train,y_train) #select classifier clf1,clf2,clf3
    prediction = clf1.predict(x_test)
    score = accuracy_score(y_test,prediction)
    f1 = f1_score(y_test , prediction ,average = 'weighted')
    
    print( 'Acurracy : ', score)
    print('F1 Score :', f1)
    
```
